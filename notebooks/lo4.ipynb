```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%config InlineBackend.figure_format = 'retina'  # for better quality plots
```

```python
import util as util
df = util.load_data()
```

# LO4: Grammar of Graphics

```python
# Now that we have the list of continents, let's create a faceted time series plot
# showing the change in life expectancy over time for each continent.

# Set the aesthetic style of the plots
sns.set_style('whitegrid')

# Create a FacetGrid with the Gapminder data, faceting by 'continent'
g = sns.FacetGrid(df, col="continent", col_wrap=3, height=4, aspect=1.5)

# Map a line plot of life expectancy over time to each facet
g = g.map(sns.lineplot, 'year', 'life_exp')

# Add titles and adjust space
g.set_titles("{col_name}")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Life Expectancy Over Time by Continent', fontsize=16)

util.save_plot(plt, 'lo4_life_expectancy_over_time_by_continent')
plt.show()
```
